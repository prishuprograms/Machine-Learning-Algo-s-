```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGLE', 'AMAZON', 'MSFT', 'FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
xpos = np.arange(len(revenue))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel('Revenue(bln)')
plt.title('Overall revenue')
plt.grid()
plt.bar(xpos - 0.2,revenue, width =0.4 ,label = "Revenue" )
plt.bar(xpos + 0.2,profit , width =0.4,label = 'Profit' )
plt.legend()
plt.plot(company,revenue, label = "Revenue", color = 'red')
plt.plot(company,profit, label = "Profit", color = 'yellow')
plt.legend()
```

```python
plt.yticks(ypos,company)

plt.title('Overall revenue')

plt.barh(xpos - 0.2,revenue,label = "Revenue" )
plt.barh(xpos + 0.2,profit, label = 'Profit' )
plt.legend()
```
